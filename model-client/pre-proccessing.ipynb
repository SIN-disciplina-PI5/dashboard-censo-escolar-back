```python
# Lendo o arquivo .csv

import pandas as pd

df = pd.read_csv('microdados_ed_basica_2022.csv', sep=';', encoding='ISO-8859-1')
```

```python
# Sample do dataset

df.head(5)
```

```python
# Enumerando as colunas do dataset

df.columns.tolist()
```

```python
# O dataset escolhido possui muitos pares de colunas de valores enumerados no formato nome/NO e código/CO,
# Então, ao invés de categorizar esses valores da coluna nome, foi necessário apenas dropar as colunas NO

# Aproveitamos também para manter apenas as colunas que serão utilizadas futuramente

colunas_manter = [
    'CO_REGIAO', 'CO_UF', 'CO_MUNICIPIO', 'CO_MESORREGIAO',
    'CO_MICRORREGIAO', 'CO_DISTRITO', 'NO_ENTIDADE', 'CO_ENTIDADE',
    'TP_DEPENDENCIA', 'TP_LOCALIZACAO', 'TP_LOCALIZACAO_DIFERENCIADA',
    'TP_SITUACAO_FUNCIONAMENTO', 'IN_VINCULO_SECRETARIA_EDUCACAO',
    'IN_VINCULO_SEGURANCA_PUBLICA', 'IN_VINCULO_SECRETARIA_SAUDE',
    'IN_VINCULO_OUTRO_ORGAO'
]

df_filtrado = df[colunas_manter]
```

```python
# Sample do dataset filtrado

df_filtrado.head(5)
```

```python
# Checando quais colunas possuem valores faltosos

df_filtrado.isnull().sum()
```

```python
# Preenchendo as lacunas na coluna TP_LOCALIZACAO_DIFERENCIADA com a moda dentre os valores

localizacao_diferenciada_mode = df_filtrado['TP_LOCALIZACAO_DIFERENCIADA'].mode()[0]
df_filtrado['TP_LOCALIZACAO_DIFERENCIADA'].fillna(localizacao_diferenciada_mode, inplace=True)
df_filtrado['TP_LOCALIZACAO_DIFERENCIADA'] = df_filtrado['TP_LOCALIZACAO_DIFERENCIADA'].astype(int)
```

```python
# As demais colunas são preenchidas com valor booleano de 0 ou 1 com caracteristicas relacionadas somente a escolas públicas,
# Então decidimos preencher os valores faltantes com o valor 2 com o significado de 'undefined'

df_filtrado['IN_VINCULO_SECRETARIA_EDUCACAO'].fillna(2, inplace=True)
df_filtrado['IN_VINCULO_SEGURANCA_PUBLICA'].fillna(2, inplace=True)
df_filtrado['IN_VINCULO_SECRETARIA_SAUDE'].fillna(2, inplace=True)
df_filtrado['IN_VINCULO_OUTRO_ORGAO'].fillna(2, inplace=True)

df_filtrado['IN_VINCULO_SECRETARIA_EDUCACAO'] = df_filtrado['IN_VINCULO_SECRETARIA_EDUCACAO'].astype(int)
df_filtrado['IN_VINCULO_SEGURANCA_PUBLICA'] = df_filtrado['IN_VINCULO_SEGURANCA_PUBLICA'].astype(int)
df_filtrado['IN_VINCULO_SECRETARIA_SAUDE'] = df_filtrado['IN_VINCULO_SECRETARIA_SAUDE'].astype(int)
df_filtrado['IN_VINCULO_OUTRO_ORGAO'] = df_filtrado['IN_VINCULO_OUTRO_ORGAO'].astype(int)
```

```python
# Sample do dataset final

df_filtrado.head(5)
```

```python
# Salvando o novo dataset

data = 'preproccessed_dataset.csv'

df_filtrado.to_csv(data, index=False)
```
